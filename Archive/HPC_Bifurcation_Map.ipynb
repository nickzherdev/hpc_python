## Write a function

```
x0 = random.uniform(0, 1)

def calc_next(r, x_prev):
    return r * x_prev * (1 - x_prev)

def plot_bif(r, steps=200, rm_first=100):
    """
    r - const
    
    """

    vals = [x0]
    x_prev = x0
    for step in range(steps):
        x_next = calc_next(r, x_prev)
        vals.append(x_next)
        x_prev = x_next
    print(len(vals[rm_first+1:]))
    plt.plot(vals[rm_first+1:])

def calc_bif(r, steps, rm_first):
    """
    r - const
    
    """

    vals = [x0]
    x_prev = x0
    for step in range(steps):
        x_next = calc_next(r, x_prev)
        vals.append(x_next)
        x_prev = x_next
    return vals[rm_first+1:]
```

## Bifurcation map


```
def plot_bif_map(start, end, resolution, steps, rm_first):

    r = np.linspace(start, end, resolution)
    X = []
    Y = []

    plt.figure(figsize=(16,8))

    for i in r:
      X.append([i]*(steps-rm_first))
      vals = calc_bif(i, steps, rm_first)
      for k in vals:
          Y.append(k)

    X = np.array(X)
    Y = np.array(Y)
    print(X.flatten().shape, Y.shape)

    plt.scatter(X.flatten(), Y, s=0.05, c='blue')

plot_bif_map(start=0, end=5, resolution=400, steps=200, rm_first=100)
```
